```python
'''
Name=G.KAVYASRI
Rollno=CS20B1126

You will generate a 3D plot similar to the one shown at

"https://matplotlib.org/stable/gallery/mplot3d/surface3d.html#sphx-glr-gallery-mplot3d-
surface3d-py" but with a different surface function of your choice. Feel free to play with

different functions (Z function in the link)
'''
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python

from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-20, 5, 0.15)
Y = np.arange(-15,3, 0.35)
X, Y = np.meshgrid(X, Y)
R = np.cbrt(5-X**2 - Y**3)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.summer,
                       linewidth=1, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.6, aspect=4)

plt.show()
```

```python
fig, bx = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(2, 5, 0.12)
Y = np.arange(-12,4, 0.37)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(9+X**4 + Y**4)
Z = np.cos(R)

# Plot the surface.
surf = bx.plot_surface(X, Y, Z, cmap=cm.autumn,
                       linewidth=1, antialiased=False)

# Customize the z axis.
bx.set_zlim(-1.01, 1.01)
bx.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
bx.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.3, aspect=7)

plt.show()
```
