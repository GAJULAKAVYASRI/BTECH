```python
'''
Name=G.KAVYASRI
Rollno=CS20B1126
4.
In this exercise, you will create a pandas data frame by reading the data from the link:
https://www.statlearning.com/s/Auto.csv.

Using the groupby command, you will calculate the average mpg for different numbers of cylinders.
You will also generate a scatter plot to show the relationship between the mpg and the displacement.
'''


import pandas as pd

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Auto.csv")
```

```python
print(df)
```

```python
# calculate the average mpg for different numbers of cylinders
df1=df.groupby("cylinders")[["mpg"]].mean()
#df1=df[["mpg","cylinders"]].groupby("cylinders").mean()
print(df1)
```

```python
#a scatter plot to show the relationship between the mpg and the displacement.
plt.scatter(df.mpg, df.displacement)
plt.title('Relationship between mpg and displacement.')
plt.xlabel('mpg')
plt.ylabel('displacement')
plt.show()
```
