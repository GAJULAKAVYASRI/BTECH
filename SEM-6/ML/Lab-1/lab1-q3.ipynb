```python
'''
Name=G.KAVYASRI
Rollno=CS20B1126

3. In this exercise, you will read an image of your choice
 using the PIL library, display it, and then
convert it to a numpy array. 
You will then work with only red plane of the array to:
a. Extract a 32x32 patch from anywhere of your choice
b. Extract the maximum pixel values along each row and column of the patch
c. Generate a new patch whose pixel values are the cosine of 
the original patch values and display it.

'''
import PIL
from PIL import Image

img = Image.open("Wallpaper.jpg")
img.show()
```

```python
#converting to numpy array
import numpy as np
mat= np.asarray(img)
print(mat)
```

```python
import matplotlib.pyplot as plt
plt.imshow(mat)
```

```python
#red plane of the array
mat1=mat[:,:,0]
plt.imshow(mat1)
```

```python
#a. Extract a 32x32 patch from anywhere of your choice
mat2=mat1[2000:2032,2000:2032]
plt.imshow(mat2)
```

```python
#b. Extract the maximum pixel values along each row and column of the patch
rowsmax=np.amax(mat2,axis=0)
columnsmax=np.amax(mat2,axis=1)
print(rowsmax)
print("\n")
print(columnsmax)
```

```python
print(mat2)
```

```python

for i  in range(32):
    max=0
    for j in range(32):
        if mat2[j,i]>max:
            max=mat2[j,i]
    print(f'Max pixel value of row{i} is {max}')
```

```python

for i  in range(32):
    max=0
    for j in range(32):
        if mat2[i,j]>max:
            max=mat2[i,j]
    print(f'Max pixel value of column{i} is {max}')
    
```

```python
#c.Generate a new patch whose pixel values are the cosine of the original patch values and display it.
matrix=np.cos(mat1)#mat1 represents red plane of the array
print(matrix)
```
