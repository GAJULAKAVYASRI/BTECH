```python
'''
Name=G.KAVYASRI
Rollno=CS20B1126

2. This exercise is designed to make you familiar with multivariate normal distribution generation
and using the generated data.
a. Generate 300 3-dimensional vectors that come from a normal distribution with
mean vector as [1 2 1]t and 3x3 covariance matrix as [5 0.8 -0.3; 0.8 3 0.6; -0.3 0.6 4]
b. Make scatter plots of x1 vs x2, x1 vs x3, and x2 vs x3. Explain whatever relationships
you can gather from these plots.
c. Calculate the mean vector and the covariance matrix using the 300 generated
points.
'''

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#a. Generate 300 3-dimensional vectors that come from a normal distribution 
m=[1,2,1]
cov=[[5,0.8,-0.3],[0.8,3,0.6],[-0.3,0.6,4]]
mt=np.transpose(m)
```

```python
x,y,z = np.random.multivariate_normal(mt,cov,300).T
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#a
print(x,y,z)
```

```python
#b. Make scatter plots of x1 vs x2, x1 vs x3, and x2 vs x3. Explain whatever relationships you can gather from these plots.
#
plt.scatter(x,y,color="blue")
plt.axis('equal')
plt.title("scatter plot between x,y")
plt.show()
```

```python
plt.scatter(x,z,color="green")
plt.axis('equal')
plt.title("scatter plot between x,z")
plt.show()
```

```python
plt.scatter(y,z,color="red")
plt.axis('equal')
plt.title("scatter plot between y,z")
plt.show()
```

```python
plt.scatter(x,y,color="blue")
plt.scatter(x,z,color="green")
plt.scatter(y,z,color="red")
plt.axis('equal')
plt.title("Comparing x,y,z")
plt.show()
```

```python
#covariance matrix
cov_mat=np.stack((x,y,z),axis=0)
print(np.cov(cov_mat))
```

```python
#mean_vector
mean_vector=(x.mean(),y.mean(),z.mean())
print(mean_vector)
```

```python
#mean_vector
print(np.mean(cov_mat,axis=1))
```
