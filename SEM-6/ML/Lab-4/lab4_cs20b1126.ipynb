```python
#Without in built function
# Gaussian Naive Bayes classifier
import csv
import numpy as np
import math
import random
```

```python
noise_level = 0.2
```

```python
#TEST CASES

d1 = [0,1,1,0,0,0,0]
d2 = [1,1,0,1,1,0,1]
d3 = [1,1,1,1,0,0,1]
d4 = [0,1,1,0,0,1,1]
d5 = [1,0,1,1,0,1,1]
d6 = [0,0,1,1,1,1,1]
d7 = [1,1,1,0,0,0,0]
d8 = [1,1,1,1,1,1,1]
d9 = [1,1,1,0,0,1,1]
d0 = [1,1,1,1,1,1,0]

digits = [d0, d1, d2, d3, d4, d5, d6,d7, d8, d9]
```

```python
def generatingDigit(digit, noise_level):
    array = []
    for segment in digit:
        rand = random.random()
        if(rand > noise_level):
            array.append(segment)
        else:
            array.append(1 - segment)
    return array
```

```python

samples =20

training = []
for d in digits:
    array = []
    for sample in range(samples):
        array.append(generatingDigit(d, noise_level))
    training.append(array)

pd.DataFrame(training).iloc[:, :]
```

```python
#Calculating probabilities

probability = []
for digit in range(10):
    arr = [0]*7
    t_set = training[:][digit]
    for row in t_set:
        for index, segment in enumerate(row):
            if(segment == 1):
                arr[index] += 1
    for index, value in enumerate(arr): #laplace smoothing #value=calculating 1's
        arr[index] = (value+1)/(2+samples)
    probability.append(arr)
df = pd.DataFrame(probability)
df.rename(columns={0: "S1", 1: "S2",2:  "S3", 3:"S4", 4: "S5", 5: "S6", 6: "S7"}, inplace=True)
print(df)
```

```python
##Testing 
ntest=5

testing = []
test_labels = []
for index, digit in enumerate(digits):
    array = []
    labels = [index]*ntest
    for example in range(ntest):
        array.append(generatingDigit(digit, noise_level))
    testing.append(array)
    test_labels.append(labels)     
pd.DataFrame(testing)
```

```python
#Prediction for text examples

#Conditional probability calculation

#max probability digit will assign 

predicting = []
for dindex, digit in enumerate(testing):
    list = []
    for example in digit:
        prob_arr = []
        for number in range(10):
            p = 1
            for seg_index, segment in  enumerate(example):
                if (segment == 1):
                    p *= probability[number][seg_index]
                else:
                    p *= (1-probability[number][seg_index])
            prob_arr.append(p)
        list.append(prob_arr.index(max(prob_arr)))
    predicting.append(list)
predicting

#predicting the digits 
```

```python
##Accuracy

X = np.array(test_labels)
Y = np.array(predicting)
accu = (X == Y).sum()/(ntest*10)
print(f"Accuracy score of our model is: {accu}")
```
